## DICIONÁRIO 2

```python
path = "https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/resources/SentiLex-lem-PT02.txt"
sentlix_dataset = pd.read_csv(path, sep=".", encoding="utf-8")
sentlix_dataset.head()
```

```python
dict_tags2 = {}
count = 0
def split_sentilex(sentence):
  global count, dict_tags2
  classe = re.split(r"[=|;]", sentence)
  classe = classe[1]
  if classe not in ["IDIOM", "V"]:
    if classe == "Adj":
      classe = "ADJ"
    else:
      classe = "NOUN"
    dict_tags2[sentlix_dataset.iloc[count][0]] = classe

  count += 1
  return sentence
```

### APLICANDO NO NOSSO DATASET DE REVIEWS DE HOTEL

```python
path_reviews = "https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/datasets/corpora/training.csv"
dataset_reviews = pd.read_csv(path_reviews, sep=",", encoding="utf-8")
dataset_reviews
```

```python
def normalize(sentence):
  new_sentence = re.sub(r"([\.|!|,|\'|\"|\(|\)|@|#|$|%|&|\*|\-|_|\+|=|\{|\}|\[|\]|~|^|:|]+)", r" \1 ", sentence)
  new_sentence = new_sentence.strip()
  new_sentence = new_sentence.lower()
  # print(new_sentence.lower())
  return new_sentence
```

```python
dataset_reviews["normalized"] = dataset_reviews.text.apply(normalize)
dataset_reviews
dataset_reviews["splitted"] = dataset_reviews.normalized.apply(str.split)
dataset_reviews
```

```python
def mapping(sentence):
  global final_dict_tags
  list_result = []
  for word in sentence:
    try:
      list_result.append(final_dict_tags[word])
    except Exception as e:
      list_result.append("X")
      # print(f"Word {word} Error:{e}")
  return list_result
```

```python
dataset_reviews["gramatical"] = dataset_reviews.splitted.apply(mapping)
dataset_reviews
```
